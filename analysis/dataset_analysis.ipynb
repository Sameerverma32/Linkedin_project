```python
import pandas as pd
import streamlit as st
import plotly.express as px
import numpy as np
```

```python
import plotly.io as pio
print(pio.templates.default)
```

```python
df = pd.read_csv("Linkedindataset.csv")
df
```

```python
df.isnull().sum()
```

```python
companyid = round(df['company_id'].mean())
print(companyid)
```

```python
df['company_id'].fillna(companyid,inplace=True)
```

```python
df.isnull().sum()
```

```python
Months_experience = round(df['months_experience'].mean())
print(Months_experience)
```

```python
df['months_experience'].fillna(Months_experience,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop('salary', axis=1, inplace=True)
df.drop('sal_low', axis=1, inplace=True)
df.drop('sal_high', axis=1, inplace=True)
df.drop('date', axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.fillna(method='ffill', inplace=True)
df.isnull().sum()
```
